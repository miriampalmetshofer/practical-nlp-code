# 01 – Web scraping

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen 
```

```python
myurl = "https://edition.cnn.com/2022/10/17/uk/liz-truss-mini-budget-apology-bbc-intl-hnk/index.html"
html = urlopen(myurl).read() 
soupified = BeautifulSoup(html, 'html.parser') 
```

As the size of the HTML webpage (soupified) is large, we are just showing some of its output (only 2000 characters).

```python
soupified.title 
```
